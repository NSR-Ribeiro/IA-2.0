```python
# Exercicio - Revisão

"""
    Qual é a sua altura?
    Qual o seu peso?
    Qual o seu IMC?
    
    Funções
        IMC = peso/altura²
"""

peso = input("Digite o seu peso: ").replace(',', '.')
peso = float(peso)

altura = input("Digite a sua altura: ").replace(',', '.')
altura = float(altura)

# Calculando o IMC
imc = peso / (altura ** 2)

# Exibindo a classificação do IMC
if imc < 17:
    print(f"Seu imc: {imc}, Está com anorexia")
elif 17 <= imc < 18.49:
    print(f"Seu imc: {imc}, Está com baixo peso")
elif 18.5 <= imc < 24.99:
    print(f"Seu imc: {imc}, Está com peso normal")
elif 25 <= imc < 29.99:
    print(f"Seu imc: {imc}, Está com sobrepeso")
elif 30 <= imc < 34.99:
    print(f"Seu imc: {imc}, Está com obesidade grau I")
elif 35 <= imc < 39.99:
    print(f"Seu imc: {imc}, Está com obesidade grau II")
else:
    print(f"Seu imc: {imc}, Está com obesidade grau III")
```

## Amanha : 28/08 tranformando um gráfico em figura

```python
plt.figure(figsize=(8,4))
# 1 polegada == 2.54 cm
# largura 8 = 20,32 cm
# altura 4 = 10.16 cm
# DPI (dots per inch, ou pontos por polegada).
# Padrão no matplotlib - DPI = 100
# Largura: 8 polegadas x 100 DPT = 800 pixels.
# Altura: 4 polegadas x 100 DPI = 400 pixels.
#plt.plot(dados_brasil['ano'],dados_brasil['imigrantes'], marker='o')
plt.plot(dados_brasil['ano'],dados_brasil['imigrantes'], lw=3, color='g')
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.yticks([500,1000,1500,2000,2500,3000])
plt.grid()
plt.title("Imigração do Brasil para o Canadá" , loc='left')
plt.xlabel("Ano")
plt.ylabel("Número de imigrantes")
plt.grid(linestyle='--')
plt.show()
```

```python
# Criar uma figura
fig,axs = plt.subplots(1,2,figsize=(15,5))

# Sublot: Gráfico de linha
axs[0].plot(dados_brasil['ano'],dados_brasil['imigrantes'])
axs[0].set_title("Imigração do Brasíl para o Canada de 1980 a 2013")
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].set_xlabel("Ano")
axs[0].set_ylabel('Número de imigrantes')

# Subplot 2: Gráfico Boxplot
axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title("Boxplot da imigração do Brasil para o canadá")
axs[1].set_xlabel("Brasíl")
axs[1].set_ylabel("Número de imigrantes")
plt.show()
```

## Comparativo de imigrações do Brasil com alguns países da America Latina

```python
# criar uma figura

fig, axs = plt.subplots(2, 2, figsize=(10, 6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)

# subplots

# Brasil
axs[0, 0].plot(df.loc['Brasil', anos])
axs[0, 0].set_title('Brasil')

# Colômbia
axs[0, 1].plot(df.loc['Colômbia', anos])
axs[0, 1].set_title('Colômbia')

# Argentina
axs[1, 0].plot(df.loc['Argentina', anos])
axs[1, 0].set_title('Argentina')

# Peru
axs[1, 1].plot(df.loc['Peru', anos])
axs[1, 1].set_title('Peru')

fig.subplots_adjust(hspace=0.5, wspace=0.6)
# O Flat simplifica esse processo ao "achatar"
# A matriz bidimensional em uma única sequência de eixos.

for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))

for ax in axs.flat:
    ax.set_xlabel('Ano', fontsize=14)
    ax.set_ylabel('Números de Imigrantes', fontsize=8)


#Usando ravel() - achatar a matriz evitando o uso compleco 
# Usando set_ylim () - definir o limite dos valores

ymin = 0
ymax = 7000
for ax in axs.ravel():
    ax.set_ylim(ymin, ymax)

# Aula 3 começa aqui
fig.suptitle("Imigração na América Latina", fontsize=18)
# Customização o Matplolib(cores, fontes, bordas)
```
