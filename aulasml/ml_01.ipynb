```python
treino_x = [papagaio1, papagaio2, papagaio3, galinha1, galinha2, galinha3, tucano1, tucano2, tucano3]
```

```python
treino_y = [1,1,1,0,0,0,2,2,2]
```

```python
# Testes
misterio= [1,1,0]
misterio1 = [1,1,0]
misterio2 = [2,1,1]
```

```python
#Previsão
teste_x = [misterio,misterio1,misterio2]
previsoes = modelo.predict(teste_x)
```

```python
teste_y = [0,1,2]
```

```python
#Função de Acurácia encpsulada 
from sklearn.metrics import accuracy_score

# Definindo uma variavel para receber o retorno da função acurácia * 100 para porcentagem
taxa_de_acerto = accuracy_score(teste_y, previsoes) * 100

#Mostrando a acurácia
print(f"Acuracia: {taxa_de_acerto:.2f}%")
```
